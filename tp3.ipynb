```python
import matplotlib.pyplot as plt
import random


# Question 1, on veut retourner les n premiers termes selon x0 et mu:

def Feigenbaum(x0, mu, n):
    x = x0
    liste = [x0]
    for i in range(n):
        x = 1 - mu * x ** 2
        liste.append(x)
    return liste
```

```python
# Considérons par exemple l'intervalle [0, 0.75] et 
# testons plusierus valeurs de x0 pour un meme mu


def testFeigenbaum(mu, n):
    for x0 in  [0.1*i for i in range(11)]:
        plt.plot(range(n+1),Feigenbaum(x0, mu, n), color='k', alpha=0.5)

    plt.title(f"Mu = {mu}, x0 entre 0 et 1")
    plt.show()

testFeigenbaum(0.5, 20)

```

On voit que ça converge pour tout x0

```python
# Considérons maintenant x0 = 0.5 et 
# testons plusieurs valeurs de mu dans cet intervalle 


def testFeigenbaum_x0(x0, n):
    for mu in  [0.1*i for i in range(7)]:
        X = range(n+1)
        Y = Feigenbaum(x0, mu, n)
        plt.plot(X,Y,  label=f"mu = {mu}")

    plt.title(f"x0 = {x0}, mu entre 0 et 0.75")
    plt.legend()
    plt.show()

testFeigenbaum_x0(0.5, 100)
```

On voit que ça converge pour toute valeur de mu

Regardons maintenant les autres intervalles

```python
liste_intervalles = [0, 0.75, 1.25, 1.368, 1.401, 2]


# 5 subfigures (3x2)
fig, axs = plt.subplots(2,3, figsize=(15, 10))

for i in range(len(liste_intervalles)-1):
    x0 = 0.5
    mu = random.uniform(liste_intervalles[i],liste_intervalles[i+1])
    axs[i//3, i%3].plot(range(51),Feigenbaum(x0, mu, 50), label=f"mu = {mu}")
    axs[i//3, i%3].set_title(f"intervalle : [{liste_intervalles[i]}, {liste_intervalles[i+1]}]")
    # axs.plot(range(51),Feigenbaum(x0, mu, 50))
    # plt.title(f"intervalle : [{liste_intervalles[i]}, {liste_intervalles[i+1]}]")

plt.legend()
plt.show()

    

```
